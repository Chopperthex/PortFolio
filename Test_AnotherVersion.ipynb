```python
import random

# Function to remove duplicate elements
def RemoveDupElements(input):
    output = [] # List for final result

    for ele in input: # Loop through each element in input[]
        if ele not in output: # If the element is first time found
            output.append(ele) # Add to output[]

    return output

# Function to select duplicate elements
def SelectDupElements(input):
    temp = [] # List for caparision
    first_half_dup = [] # List for first found duplicate
    second_half_dup = [] # List for second half duplicate
    output = [] # List for final result

    for ele in input: # Loop through each element in input[]

        if ele not in temp: # If the element is first time found
            temp.append(ele) # Add to temp[]
        else: # Element is already in temp[] = Duplicated element
            second_half_dup.append(ele) # Add to second half list

    for nums in temp: # Loop through each element in temp[]
        if nums in second_half_dup: # Find first found duplicate
            first_half_dup.append(nums) # Add to first half list

    output = first_half_dup + second_half_dup # Combine list

    return output

#TESTING
def testRemoveDupElements(): # RemoveDupElements testing

    pass_case = 0 # Passed case counter
    fail_case = 0 # Failed case counter
    round = 0 # Testing Round
    failure_position = [] # Failure pointer

    for round in range(100): #100 times
        print("Round :", round+1) # Round Counting
        print("-----------------------------------------------------------------")
        random_input = [random.randint(1, 9) for i in range(7)] # Random 7 integers from 1-9 in the list
        expected_output = [] # Result

        for ele in random_input: # Loop through each element in input list
            if ele not in expected_output: # If the element is first time found
                expected_output.append(ele) # Add to testing output[]

        result = RemoveDupElements(random_input) # Result from actual function

        print(random_input)
        print(expected_output)

        if result == expected_output : # Work as expected
            print("[1] : PASS")
            pass_case += 1
        else: # Failed
            print("[1] : FAILED")
            fail_case += 1
            failure_position.append(round+1)
        print("-----------------------------------------------------------------")

    # Summary
    print("Test Case: ", round+1)
    print("Success: ", pass_case)
    print("Failure: ", fail_case)
    print("Failure Round: ", failure_position)
    print("-----------------------------------------------------------------")

def testSelectDupElements(): # SelectDupElements testing

    pass_case = 0 # Passed case counter
    fail_case = 0 # Failed case counter
    round = 0 # Testing Round
    failure_position = [] # Failure pointer

    for round in range(100):
        print("Round :", round+1) # Round Counting
        random_input = [random.randint(1, 9) for i in range(7)] # Random 7 integers from 1-9 in the list
        temp = [] # Comparision list
        first_half_dup = [] # List for first found duplicate
        second_half_dup = [] # List for second half duplicate
        expected_output = first_half_dup + second_half_dup # Result

        for ele in random_input: # Loop through each element in input list
            if ele not in temp: # If the element is not already in output
                temp.append(ele) # Add to temp[]
            else: # Duplicated element
                second_half_dup.append(ele) # Add to second half list

        for nums in temp: # Loop through each element in temp[]
            if nums in second_half_dup: # Find first found duplicate
                first_half_dup.append(nums)

        expected_output = first_half_dup + second_half_dup # Combine list
        result = SelectDupElements(random_input) # Result from actual function

        print(random_input)
        print(expected_output)

        if result == expected_output : # Work as expected
            print("[2] : PASS")
            pass_case += 1
        else: # Failed
            print("[2] : FAILED")
            fail_case += 1
            failure_position.append(round+1)
        print("-----------------------------------------------------------------")

    # Summary
    print("Test Case: ", round+1)
    print("Success: ", pass_case)
    print("Failure: ", fail_case)
    print("Failure Round: ", failure_position)
    print("-----------------------------------------------------------------")

if __name__ == '__main__':
    testRemoveDupElements()
    testSelectDupElements()
```
